```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Day_14_Pharma_data.csv')

print("Missing values in each column:")
print(df.isnull().sum())

duplicates = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")

df = df.drop_duplicates()

df = df.dropna()

print("\nCleaned dataset:")
print(df.head())
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='Product_Name', y='Effectiveness', hue='Region', data=df, palette='viridis')
plt.title('Average Effectiveness for Each Drug Across Different Regions')
plt.xlabel('Product')
plt.ylabel('Average Effectiveness')
plt.legend(title='Region')
plt.show()
```

```python
plt.figure(figsize=(14, 10))
sns.violinplot(x='Product_Name', y='Effectiveness', data=df, palette='viridis')
plt.title('Distribution of Effectiveness for Each Product')
plt.xlabel('Product')
plt.ylabel('Effectiveness')
plt.show()

plt.figure(figsize=(14, 10))
sns.violinplot(x='Product_Name', y='Side_Effects', data=df, palette='viridis')
plt.title('Distribution of Side Effects for Each Product')
plt.xlabel('Product')
plt.ylabel('Side Effects')
plt.show()
```

```python
sns.pairplot(df[['Effectiveness', 'Side_Effects', 'Marketing_Spend']], palette='viridis')
plt.suptitle('Pairplot of Effectiveness, Side Effects, and Marketing Spend', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Trial_Period', y='Effectiveness', data=df, palette='viridis')
plt.title('Effectiveness for Different Trial Periods')
plt.xlabel('Trial Period')
plt.ylabel('Effectiveness')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.regplot(x='Marketing_Spend', y='Effectiveness', data=df, scatter_kws={'s':10}, line_kws={'color':'red'})
plt.title('Effect of Marketing Spend on Drug Effectiveness')
plt.xlabel('Marketing Spend')
plt.ylabel('Effectiveness')
plt.show()
```
