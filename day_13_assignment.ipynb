```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Day_13_Pharma_data.csv')

print("Missing values in each column:")
print(df.isnull().sum())

duplicates = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")

df = df.drop_duplicates()

df = df.dropna()

print("\nCleaned dataset:")
print(df.head())
```

```python
total_sales_per_region = df.groupby('Region')['Sales'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='Sales', data=total_sales_per_region, palette='viridis')
plt.title('Total Sales per Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Marketing_Spend', y='Sales', data=df, hue='Region', palette='viridis')
plt.title('Relationship between Marketing Spend and Sales')
plt.xlabel('Marketing Spend')
plt.ylabel('Sales')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age_Group', y='Effectiveness', data=df, palette='viridis')
plt.title('Drug Effectiveness across Different Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Effectiveness')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(x='Trial_Period', y='Sales', hue='Product_Name', data=df, marker='o', palette='viridis')
plt.title('Sales Trend for Each Product over Different Trial Periods')
plt.xlabel('Trial Period')
plt.ylabel('Sales')
plt.show()
```

```python
correlation_matrix = df[['Sales', 'Marketing_Spend', 'Effectiveness']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt='.2f')
plt.title('Correlation between Sales, Marketing Spend, and Effectiveness')
plt.show()
```
